```python
#i) cantidad de vacunas utilizadas cada día
import pandas as pd
Vacunas = pd.read_csv("/content/drive/MyDrive/Data Science/COVID-19/country_vaccinations.csv")
Vacunas['date']=pd.to_datetime(Vacunas['date'])
conCantidad = Vacunas
conCantidad['canVac']= conCantidad['vaccines'].str.split(',')
conCantidad[['canVac','total_vaccinations']].groupby(conCantidad['date'].dt.day).sum()
```

```python
#k) registros donde se haya utilizado la vacuna Pfizer.
import pandas as pd
Vacunas = pd.read_csv("/content/drive/MyDrive/Data Science/COVID-19/country_vaccinations.csv")
pfizer = Vacunas
pfizer[pfizer['vaccines'].str.contains('Pfizer', case=False)]
```

```python
#i) Recrear los dataframes anteriormente generados (conDiferencias, conCantidad, antes20 y pfizer) y guardar un archivo de Excel denominado resultadosReto.xlsx
import pandas as pd
Vacunas = pd.read_csv("/content/drive/MyDrive/Data Science/COVID-19/country_vaccinations.csv")
Vacunas['date']=pd.to_datetime(Vacunas['date'])
#Pfizer Dataframe
pfizer = Vacunas
pfizer[pfizer['vaccines'].str.contains('Pfizer', case=False)]
#Antes20 Dataframe
antes20 = Vacunas
antes20.query("date<'2020-12-20'")
#conCantidad Dataframe
conCantidad = Vacunas
conCantidad['canVac']= conCantidad['vaccines'].str.split(',')
conCantidad[['canVac', 'date', 'total_vaccinations']].groupby(conCantidad['date'].dt.day).sum()
# Export to excel
xls1 = pd.DataFrame([[conCantidad]])
xls2 = pd.DataFrame([[antes20]])
xls3 = pd.DataFrame([[pfizer]])
with pd.ExcelWriter("/content/drive/MyDrive/Data Science/resultadosReto.xlsx", mode='w') as writer:
  xls1.to_excel(writer, sheet_name="conCantidad")
  xls2.to_excel(writer, sheet_name="Antes20")
  xls3.to_excel(writer, sheet_name="Pfizer")
```
